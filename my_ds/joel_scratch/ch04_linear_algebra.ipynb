## dot product
The dot product `def dot(v: Vector, w: Vector) -> float:` measures how far the vector *v* extends in the *w* direction. For example, if `w = [1, 0]` then `dot(v, w)` is just the first component of *v*. Another way of saying this is that it’s the length of the vector you’d get if you *projected* *v* onto *w*.

The dot product of two vectors is the sum of their componentwise products.

Perpendicular vectors have a dot product equal to zero.

check this

```python
from typing import List
Vector = List[float]

# dot product. The dot product of two vectors is the sum of their componentwise products
def dot(v: Vector, w: Vector) -> float:
    """sum of the component wise product of two vectors
    v_1 * w_1 + ... + v_n * w_n

    The dot product measures how far the vector v extends in the w direction.
    For example, if w = [1, 0] then dot(v, w) is just the first component of v.
    Another way of saying this is that it’s the length of the vector you’d get
    if you projected v onto w.
    """

    # sum dimension/element wise
    return sum(v_i * w_i for v_i, w_i in zip(v, w))
```
