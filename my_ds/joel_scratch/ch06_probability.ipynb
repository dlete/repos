## Dependence and Independence

### Independence

Roughly speaking, we say that two events E and F are *dependent* if knowing something about whether E happens gives us information about whether F happens (and vice versa). Otherwise they are *independent*.

Events E and F are independent if the probability that they both happen is the product of the probabilities that each one happens

```bash
P(E,F) = P(E)P(F)
```


## Conditional probability

Independent events

$P(E,F) = P(E)P(F)$

Non-independent events. Probability of E “conditional on F”. Think of it as the probability that E happens, given that we know that F happens.

$P(E|F) = P(E,F)/P(F)$ can also be rewritten as $P(E,F) = P(E|F) P(F)$

## Random Variables

A ***random variable*** is a variable whose possible values have an associated probability distribution.

**Random Variable => probability distribution**

Example: 
* the probability distribution for the coin flip variable equals 0 with probability 0.5 and 1 with probability 0.5.
* the probability distribution for the range(10) variable has a distribution that assigns probability 0.1 to each of the numbers from 0 to 9.

The sum of all the values in the probability distribution add to 1. If the probability distribution is continuous, then the integral of the probability distribution is (adds) 1.

The ***expected value*** of a random variable is the average of its values weighted by their probabilities. The coin flip variable has an expected value of 1/2 (= 0 * 1/2 + 1 * 1/2), and the range(10) variable has an expected value of $4.5 = 0 * 0.1 + 1 * 0.1 + 2 * 0.1 + 3* 0.1  + 4* 0.1  + 5* 0.1  + 6* 0.1  + 7* 0.1  + 8* 0.1  + 9* 0.1$.


````python

def expected_value(xs: List[float], p: float) -> float:
    """retuns expected value for a data set of events with a given probability for each event"""
    ev = sum([x * p for x in xs])
    return ev
    
# expected value for flip coin
coin_values = [0, 1]
coin_probability = 0.5
outcome1 = expected_value(coin_values, coin_probability)
print("the expected_value of coin flip values", coin_values, "with probability", coin_probability, "is:", outcome1)

# expected value of range(10) when each element has 0.1 proability
range10_values = range(10)
range10_probability = 0.1
outcome2 = expected_value(range10_values, range10_probability)
print("the expected_value of range10 values values", range10_values, "with probability", range10_probability, "is:", round(outcome2, 2))
````

## Continuous Distributions

We represent a ***continuous distribution*** with a ***probability density function*** (pdf) such that the probability of seeing a value in a certain interval equals the integral of the density function over the interval.

The ***cumulative distribution function*** (cdf) gives the probability that a random variable is less than or equal to a certain value.

````python
#
# Graph the uniform pdf
#
# x-axis and y-axis values
xs = [x / 10.0 for x in range(-10, 20)]
ys = [uniform_pdf(x) for x in xs]

# produce a plot graph
plt.plot(xs, ys)

# add title, and labels, and show
plt.title("The uniform pdf")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()


#
# Graph the uniform cdf
#
# x-axis and y-axis values
xs = [x / 10.0 for x in range(-10, 20)]
ys = [uniform_cdf(x) for x in xs]

# produce a plot graph
plt.plot(xs,ys)

# add title, and labels, and show
plt.title("The uniform cdf")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()
````

## The Normal Distribution

It is the classic bell curve–shaped distribution and is completely determined by two parameters: its ***mean μ (mu)*** and its ***standard deviation σ (sigma)***. The mean indicates where the bell is centered, and the standard deviation how “wide” it is.

When *μ = 0* and *σ = 1*, it’s called the ***standard normal distribution***.


$f(x \mid \mu, \sigma) = \frac {1}{\sqrt {2 \pi} \sigma} exp \left( - \frac{(x - \mu)^2}{2 \sigma^2} \right)$ 

````python
#
# Graph the Normal Distribution PDF for different mu, sigma
#
# x-axis values
xs = [x / 10.0 for x in range(-50, 50)]
ys = [normal_pdf(x,sigma=1) for x in xs]

# produce plot graphs
plt.plot(xs, ys,'-',label='mu=0,sigma=1')

# add title and labels to the x-axis and y-axis
plt.legend(loc=1) # upper right
plt.title("PDF for the Normal Distirbution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()
````

````python
#
# Graph the Normal Distribution PDF for different mu, sigma
#
# x-axis values
xs = [x / 10.0 for x in range(-50, 50)]

# produce plot graphs
plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
plt.plot(xs,[normal_pdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
plt.plot(xs,[normal_pdf(x,mu=-1) for x in xs],'-.',label='mu=-1,sigma=1')

# add title and labels to the x-axis and y-axis
plt.legend(loc=1) # upper right
plt.title("Various Normal pdfs")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()


#
# Graph the Normal Distribution CDF for different mu, sigma
#
# x-axis values
xs = [x / 10.0 for x in range(-50, 50)]

# produce plot graphs
plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
plt.plot(xs,[normal_cdf(x,mu=-1) for x in xs],'-.',label='mu=-1,sigma=1')

# add title and labels to the x-axis and y-axis
plt.legend(loc=4) # lower right
plt.title("Various Normal cdfs")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()


def inverse_normal_cdf(p: float,
                       mu: float = 0,
                       sigma: float = 1,
                       tolerance: float = 0.00001) -> float:
    """find approximate inverse using binary search

    Find the value corresponding to a specified probability
    Input probability -> return/output value with that probability
    """

    # if not standard, compute standard and rescale
    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)

    low_z = -10.0                      # normal_cdf(-10) is (very close to) 0
    hi_z  =  10.0                      # normal_cdf(10)  is (very close to) 1
    while hi_z - low_z > tolerance:
        mid_z = (low_z + hi_z) / 2     # Consider the midpoint
        mid_p = normal_cdf(mid_z)      # and the cdf's value there
        if mid_p < p:
            low_z = mid_z              # Midpoint too low, search above it
        else:
            hi_z = mid_z               # Midpoint too high, search below it

    return mid_z
````

````python
one_sigma_pos = normal_cdf(1)
one_sigma_neg = normal_cdf(-1)
print('normal CDF of (one sigma - (-one sigma)) is:', round(one_sigma_pos - one_sigma_neg, 2), 'percent')

two_sigma_pos = normal_cdf(2)
two_sigma_neg = normal_cdf(-2)
print('normal CDF of (two sigma - (-two sigma)) is:', round(two_sigma_pos - two_sigma_neg, 2), 'percent')

three_sigma_pos = normal_cdf(3)
three_sigma_neg = normal_cdf(-3)
print('normal CDF of (three sigma - (-three sigma)) is:', round(three_sigma_pos - three_sigma_neg, 3), 'percent')
````

## The Central Limit Theorem

The ***central limit theorem***, which says (in essence) that a random variable defined as the average of a large number of independent and identically distributed random variables is itself approximately normally distributed.

The mean of a Bernoulli(p) variable is *p*, and its standard deviation is *sqr(p(1 − p))*.

The central limit theorem says that as n gets large, a Binomial(n,p) variable is approximately a normal random variable with mean *μ = np* and standard deviation *σ = sqr(np(1 − p )*.

````python

def make_hist(n: int, p: float, num_points: int) -> plt.Figure:
    """creates a composite graph: histogram for binonial samples and a line for the normal approximation
    
    n, independent Bernoulli(p) random variables
    p, probability
    """
    # binomial samples
    data = [binomial(n, p) for _ in range(num_points)]
    histogram = Counter(data)

    # binomial samples, data, x-axis and y-axis values
    xs_binomial = [x - 0.4 for x in histogram.keys()]
    ys_binomial = [v / num_points for v in histogram.values()]

    # binomial samples, graph, use a bar chart 
    plt.bar(xs_binomial, ys_binomial, 0.8, color='0.75')    


    # normal approximiation, data, x-axis and y-axis
    xs_normal = range(min(data), max(data) + 1)
    mu = p * n
    sigma = math.sqrt(n * p * (1 - p))
    ys_normal = [normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma)
        for i in xs_normal]

    # normal approximiation, graph using a line chart
    plt.plot(xs_normal, ys_normal)

    # add title, add labels, and show
    plt.title("Binomial Distribution vs. Normal Approximation")
    plt.xlabel("Values")
    plt.ylabel("Probability")
    plt.show()


# graph
make_hist(100, 0.75, 10000)
````
