### Central tendencies

A generalization of the *median* is the *quantile*, which represents the value less than which a certain percentile of the data lies. (The *median* represents the value less than which 50% of the data lies.)

Is the "index" value or the actual "xs" value? -> the **xs** value

```python

try:
    outcome = mean(num_friends)   # 7.33
    print("the 'mean' of data set num_friends is float: " + str(outcome))
except Exception as err:
    print("found exception in the function 'median': " + str(err))


try:
    outcome = median(num_friends)           # 6
    print("the 'median' of data set num_friends is float: " + str(outcome))
except Exception as err:
    print("found exception in the function 'median': " + str(err))


    
try:
    outcome10 = quantile(num_friends, 0.10)
    outcome25 = quantile(num_friends, 0.25)
    outcome75 = quantile(num_friends, 0.75)
    outcome99 = quantile(num_friends, 0.99)
    print("the 10th percentile, the 10 quantile, of num_friends is float: " + str(outcome10))  # 1
    print("the 25th percentile, the 25 quantile, of num_friends is float: " + str(outcome25))  # 3
    print("the 75th percentile, the 75 quantile, of num_friends is float: " + str(outcome75))  # 9
    print("the 99th percentile, the 99 quantile, of num_friends is float: " + str(outcome99))  # 41
except Exception as err:
    print("found exception in the function 'quantile': " + str(err))


try:
    outcome = mode(num_friends)     # [6, 1]
    print("the 'modes' of num_friends are the list: " + str(outcome))  # [6, 1]
except Exception as err:
    print("found exception in the function 'mode': " + str(err))
```

### Dispersion

#### Variance

$variance = \sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n - 1}$

When we’re dealing with a ***sample*** from a larger *population*, *x_bar* is only an estimate of the actual mean, which means that on average *(x_i - x_bar) ** 2* is an underestimate of *x_i*’s squared deviation from the mean, which is why we divide by *n-1* instead of *n*. See [Wikipedia](http://bit.ly/1L2EapI).

The *variance* has units that are the square of the original units (e.g., “friends squared”). As it can be hard to make sense of these, we often look instead at the *standard deviation*: $\sigma = \sqrt{variance}$

A more robust alternative, the *interquartile range*, computes the difference between the 75th percentile value and the 25th percentile value:

```python

try:
    outcome = data_range(num_friends)           # 99
    print("the 'range' of num_friends is float: ", outcome)
except Exception as err:
    print("found exception in the function 'data_range': ", err)


try:
    outcome = variance(num_friends)             # 81.54
    print("the 'variance' of num_friends is float: ", outcome)
except Exception as err:
    print("found exception in the function 'variance': ", err)


try:
    outcome = standard_deviation(num_friends)   # 9.03
    print("the 'standard_deviation' of num_friends is float: ", outcome)
except Exception as err:
    print("found exception in the function 'standard_deviation': ", err)


try:
    outcome = interquartile_range(num_friends)  # 6
    print("the 'interquartile_range' of num_friends is float: ", outcome)
except Exception as err:
    print("found exception in the function 'interquartile_range': ", err)
```

## Correlation

### Covariance

We’ll first look at ***covariance***, the paired analogue of variance. Whereas variance measures how a single variable deviates from its mean, covariance measures how two variables vary in tandem from their means.

Nonetheless, this number (covariance that is) can be hard to interpret, for a couple of reasons:
* Its units are the product of the inputs’ units (e.g., friend-minutes-per-day), which
can be hard to make sense of. (What’s a “friend-minute-per-day”?)
* If each user had twice as many friends (but the same number of minutes), the
covariance would be twice as large. But in a sense the variables would be just as
interrelated. Said differently, it’s hard to say what counts as a “large” covariance.

### Correlation

For this reason, it’s more common to look at the ***correlation***, which divides out the
standard deviations of both variables.

The correlation is unitless and always lies between -1 (perfect anti-correlation) and 1 (perfect correlation). A number like 0.25 represents a relatively weak positive correlation.

```python

# graph scatter plot
plt.scatter(num_friends, daily_minutes)

# title, legends and show
plt.title("Correlation with one outliner (daily minutes vs. # of friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()

outcome = correlation(num_friends, daily_minutes)       # 0.25
print("the 'correlation for num_friends and daily_minutes is float: ", outcome)
```

```python
# The person with 100 friends (who spends only one minute per day on the site) is a huge outlier, and correlation can be very sensitive to outliers. What happens if we ignore him?

# index of outlier
outlier = num_friends.index(100)

num_friends_good = [x
                    for i, x in enumerate(num_friends)
                    if i != outlier]

daily_minutes_good = [x
                      for i, x in enumerate(daily_minutes)
                      if i != outlier]



# graph scatter plot
# plt.axis("equal")     # sets both axes to be equal
plt.scatter(num_friends_good, daily_minutes_good)

# title, legend and show
plt.title("Correlation after romoving outliner (daily minutes vs. # of friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()


# Without the outlier, there is a much stronger correlation
outcome = correlation(num_friends_good, daily_minutes_good) # 0.57
print("the 'correlation for num_friends_good and daily_minutes_good is float: ", outcome)
```

## Simpson’s Paradox

One not uncommon surprise when analyzing data is Simpson’s Paradox, in which correlations can be misleading when ***confounding*** variables are ignored.

This phenomenon crops up in the real world with some regularity. The key issue is that correlation is measuring the relationship between your two variables ***all else being equal***. If your data classes are assigned at random, as they might be in a well designed experiment, *“all else being equal”* might not be a terrible assumption. But when there is a deeper pattern to class assignments, *“all else being equal”* can be an awful assumption.

The only real way to avoid this is by knowing your data and by doing what you can to
make sure you’ve checked for possible confounding factors. 

Remember the example: imagine that you can identify all of your members as either East Coast data scientists or West Coast data scientists. You decide to examine which coast’s data scientists are friendlier.

## Some Other Correlational Caveats

A correlation of zero indicates that there is no *linear* relationship between the two variables. However, there may be other sorts of relationships. 

## Correlation and Causation
This is an important point: if x and y are strongly correlated, that might mean that x causes y, that y causes x, that each causes the other, that some third factor causes both, or it might mean nothing.

One way to feel more confident about causality is by conducting ***randomized*** trials. If you can randomly split your users into two groups with similar demographics and give one of the groups a slightly different experience, then you can often feel pretty good that the different experiences are causing the different outcomes.

For instance, if you don’t mind being angrily accused of experimenting on your users, you could randomly choose a subset of your users and show them content from only a fraction of their friends. If this subset subsequently spent less time on the site, this would give you some confidence that having more friends causes more time on the site.
